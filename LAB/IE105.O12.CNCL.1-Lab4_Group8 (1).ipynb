**3. Sinh viên tự tìm hiểu thực hiện lại ví dụ dùng mô hình Linear Regression trong thư viện scikit-learn bằng các thư viện sau:**

Cho biết cảm nghĩ về việc dùng 4 thư viện này

*Dùng mô hình Linear Regression trong thư viện scikit-learn:*

```python
%%time
from sklearn.linear_model import LinearRegression

X = np.array([[3], [5], [7], [9], [11]]).reshape(-1, 1)
y = [8.0, 9.1, 10.3, 11.4, 12.6]
test = np.array([[13]])

# Create a linear regression model
model_sklearn = LinearRegression()

# Fit the model to the training data
model_sklearn.fit(X, y)

# Predict using the test data
prediction_sklearn = model_sklearn.predict(test.reshape(-1, 1))

print("Scikit-learn prediction:", prediction_sklearn)
```

*Dùng mô hình Linear Regression trong thư viện TensorFlow:*

```python
%%time
import numpy as np
import tensorflow as tf

# Data
X_tf = np.array([[3], [5], [7], [9], [11]]).reshape(-1, 1)
y_tf = np.array([8.0, 9.1, 10.3, 11.4, 12.6]).reshape(-1, 1)
test = np.array([[13]])

# Convert data to TensorFlow constants
X_tf = tf.constant(X, dtype=tf.float32)
y_tf = tf.constant(y, dtype=tf.float32)
test_tf = tf.constant(test, dtype=tf.float32)

# Variables
W = tf.Variable(tf.random.normal([1, 1]), name='weight')
b = tf.Variable(tf.zeros([1]), name='bias')

# Linear regression model
def linear_regression(X):
    return tf.matmul(X, W) + b

# Mean squared error loss function
def mean_squared_error(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))

# Optimizer
learning_rate = 0.01
optimizer = tf.optimizers.SGD(learning_rate)

# Training
epochs = 1000
for epoch in range(epochs):
    # Use tf.GradientTape to record the operations for automatic differentiation
    with tf.GradientTape() as tape:
        predictions = linear_regression(X_tf)
        loss = mean_squared_error(y_tf, predictions)

    # Compute gradients
    gradients = tape.gradient(loss, [W, b])

    # Update weights and bias
    optimizer.apply_gradients(zip(gradients, [W, b]))

# Predict using the test data
prediction_tf = linear_regression(test_tf)

print("TensorFlow prediction:", prediction_tf.numpy()[0][0])
```

*Dùng mô hình Linear Regression trong thư viện Keras:*

```python
%%time
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers

# Data
X = np.array([[3], [5], [7], [9], [11]]).reshape(-1, 1)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6])

# Build the model using Keras Sequential API
model = keras.Sequential([
    layers.Input(shape=(1,)),
    layers.Dense(units=1)
])

# Compile the model
optimizer = keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Fit the model to the training data
model.fit(X, y, epochs=1000, verbose=0)

# Predict using the test data
prediction_keras = model.predict(np.array([[13]]))

print("Keras prediction:", prediction_keras[0][0])
```

*Dùng mô hình Linear Regression trong thư viện PyTorch:*

```python
%%time
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# Data
X = np.array([[3], [5], [7], [9], [11]], dtype=np.float32)
y = np.array([8.0, 9.1, 10.3, 11.4, 12.6], dtype=np.float32).reshape(-1, 1)
test = np.array([[13]], dtype=np.float32)

# Convert data to PyTorch tensors
X_tensor = torch.from_numpy(X)
y_tensor = torch.from_numpy(y)
test_tensor = torch.from_numpy(test)

# Define the model
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super(LinearRegressionModel, self).__init__()
        self.linear = nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

# Instantiate the model
model = LinearRegressionModel()

# Define loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training
epochs = 1000
for epoch in range(epochs):
    # Forward pass
    predictions = model(X_tensor)
    loss = criterion(predictions, y_tensor)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

# Predict using the test data
prediction_pytorch = model(test_tensor)

print("PyTorch prediction:", prediction_pytorch.item())
```
